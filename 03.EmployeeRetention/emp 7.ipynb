### Explore the distributions of length of employment and weeks of year when employees quit

```python
length=pd.to_datetime(df['quit_date'])-pd.to_datetime(df['join_date'])
```

```python
length1=length.apply(lambda x:float(x.days))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('Employment Length')
ax.set_ylabel('Count')
ax.hist(length1[np.isfinite(length1)],bins=100)
ax.axvline(x=500,color='red')
ax.axvline(x=396,color='orange')
ax.axvline(x=250,color='yellow')
plt.show()
```

Around every year (~365 days, 730 days), there are peaks in the number of employees who quit. It seems that people are more likely to quit close to their anniversary. When they get a bonus?

Additionally, it looks like hardly anyone quits before about 250 days (yellow line). Then the peak quitting period happens around 13 months (396 days, orange line). Finally, the first 'group' of quitting lasts until about 500 days/1 year 3 months (red line).

```python
week_of_year = pd.to_datetime(df['quit_date'])
week_of_year = week_of_year.apply(lambda x:float(x.week))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlabel('Week of Year when Quit')
ax.set_ylabel('Count')
ax.hist(week_of_year[np.isfinite(week_of_year)],bins=50)
plt.show()
```

It looks like people are quitting around New Year's. Maybe because they get bonuses then and because it is easier to find a new job since companies have more money to hire at the beginning of the year?

### Prediction with decision tree: predict early quitters to understand why they leave

First I will define an early quitter as someone who leaves within 13 months (from the exploratory plot above). Then I will define an early quitting as someone who leaves within 1 year and 3 months (from the exploratory plot above). I will predict an early quitter rather than predicing whether someone is still at the company, so that I can identify what causes someone to quit early, and so that the company can take steps to prevent those people from leaving.

```python
def is_employed(row):
    if row['length'] < 396 and row['quit_date'] == row['quit_date']:
        val = 1
    else:
        val = 0
    return val
df['employed'] = df.apply(is_employed,axis=1)
df.head()
```

```python
df_for_tree = df[pd.to_datetime(df['join_date'])<pd.to_datetime('2015-12-13')-pd.to_timedelta(396,unit='d')]
```

```python
X_train, X_test, y_train, y_test = train_test_split( \
df_for_tree[['week_of_year','company_id','customer_service','data_science','design','engineer','marketing','sales','seniority','salary']], df_for_tree['employed'], test_size=0.33, random_state=4)
```

```python
clf = tree.DecisionTreeClassifier()

clf = clf.fit(X_train, y_train)
predicted = clf.predict(X_test)
```

Adding week of year when quit gives about 5% increase in accuracy when not including it.

```python
accuracy_score(predicted, y_test)
```

For the model with week ofyear information, the most important feature is now week of year, followed by salary. Though this might not be too useful as I'm not sure how much a company could do to prevent people from leaving at a certain time of year.

```python
clf.feature_importances_
```
